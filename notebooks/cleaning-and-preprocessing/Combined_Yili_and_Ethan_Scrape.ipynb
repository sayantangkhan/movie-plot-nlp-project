```python
csv_input = pd.read_csv('drive/MyDrive/wiki_movie_plots_deduped_c2.csv')
csv_input = csv_input.dropna() #gets rid of rows with None value 
```

```python
combined_1 = pd.read_csv('/content/drive/My Drive/imdb_temporary_output_combined.csv')
```

```python
combined_2 = pd.read_csv('imdb_temporary_output_yili.csv')
```

```python
combined_2_list = copy.copy(list(combined_2.imdb_summaries.values))
combined_1_list = copy.copy(list(combined_1.imdb_summaries.values))
```

```python
combined_2_list.append("!!!DELETE ME!!!") # for the extra missing entry
```

```python
for i in range(len(combined_2_list)):
  entry = combined_1_list[i]
  update = False
  if entry == '' or entry == ' ' or entry == "!!!DELETE ME!!!" or entry == 'nan':
    update = True
  new_entry = combined_2_list[i]
  if new_entry == "!!!DELETE ME!!!" or new_entry == 'nan':
    combined_2_list[i] = ''
  elif new_entry != '' and new_entry != ' ':
    if update:
      combined_1_list[i] = copy.copy(combined_2_list[i])
      combined_2_list[i] = ''
  else:
    if update:
      combined_1_list[i] = ''
```

```python
for k in combined_1_list:
  print(k)
for l in combined_2_list:
  print(l)
```
