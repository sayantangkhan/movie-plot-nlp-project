```python
import nltk
nltk.download('punkt')
```

```python
def decompose_paragraph(paragraph):
  sentences = nltk.tokenize.sent_tokenize(paragraph)
  if len(sentences) == 0:
    return []
  
  paragraph_chunks = []
  paragraph_chunks.append(sentences[0])
  for sentence in sentences[1:]:
    #print(len((paragraph_chunks[-1] + " " + sentence).split()))
    if len((paragraph_chunks[-1] + " " + sentence).split()) <= 256:
       paragraph_chunks[-1] = paragraph_chunks[-1] + sentence
    else:
      paragraph_chunks.append(sentence)
  return paragraph_chunks

def decompose_plot(plot):
  paragraphs = plot.split('\r\n')
  chunks = []
  for paragraph in paragraphs:
    chunks.extend(decompose_paragraph(paragraph))
  return chunks
```

```python
plots['to_embed'] = plots.apply(lambda row: decompose_plot(row['Plot']), axis=1)
```

```python
plots.to_csv("../../../imdb_plots.csv", index=False, compression="zip")
```
